```python
MOD_LETTERS = ["A", "B", "C"]
WEAPON_URLS = {}

# With overclocks
for weapon, attr in weapons.items():
    # Create build
    id = attr.get("id")
    mod_slots = [MOD_LETTERS[:tier] for tier in attr.get("mod_slots")]
    mod_combinations = ["".join(c) for c in list(itertools.product(*mod_slots))]
    for mods in mod_combinations:
        for overclock_id, overclock in enumerate(attr.get("overclocks")):
            name = f"{weapon} ({mods}) + {overclock}"
            url = f"https://karl.gg/asv/{id}/{mods}{overclock_id+1}#/"
            WEAPON_URLS[name] = url
            
# Base builds
for weapon, attr in weapons.items():
    name = f"{weapon} (Base)"
    url = f"https://karl.gg/asv/{attr.get('id')}/------#/"
    WEAPON_URLS[name] = url
```

```python
def create_table(weapon_urls):
    dfs = []
    for weapon_name, url in tqdm(weapon_urls.items()):
        #print(url)
        burst, sust, misc = pd.read_html(url)   
        burst_df = burst.set_index(0).transpose().add_suffix(" (Burst)")
        sust_df = sust.set_index(0).transpose().add_suffix(" (Sustained)")
        misc_df = misc.set_index(0)
        df = pd.concat([burst_df, sust_df, misc_df], axis=1)
        df["name"] = weapon_name
        dfs.append(df)
    return dfs
```

```python
all_df = create_table(WEAPON_URLS)
```
